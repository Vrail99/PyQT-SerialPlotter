
Gleichung für Venturirohr

Die Druckdifferenz im Venturirohr ergibt sich aus der Bernoulli-Gleichung:

$$
p_1 - p_2 = \frac{Q^2 \cdot \rho}{2} \left( \frac{1}{A_2^2} - \frac{1}{A_1^2} \right)
$$

Dabei ist  
$p_1, p_2$ – Druck vor und nach der Verengung  
$Q$ – Volumenstrom  
$\rho$ – Dichte des Mediums  
$A_1, A_2$ – Querschnittsflächen vor und nach der Verengung
